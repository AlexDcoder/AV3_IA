## Instalando dependências

```python
import matplotlib.pyplot as plt
import numpy as np
from src.learning import Data, ModelsRNA, Validation

data = Data('spiral.csv')
```

## Gerando gráfico inicial

```python
# Gerando gráfico com base nas informações
plt.title(f"Gráfico gerado à partir dos dados do {data.file_name}.")

for category in data.categories:
    # Desenhar pontos no gráficos
    plt.scatter(data.X[0, data.Y == category],
                data.X[1, data.Y == category],  edgecolors='k',
                label=f'Categoria {int(category)}')

# Definir limites visuais dos eixos x e y
plt.xlim(np.min(data.X[0, :]) - .5, np.max(data.X[0, :]) + .5)
plt.ylim(np.min(data.X[1, :]) - .5, np.max(data.X[1, :]) + .5)

plt.xlabel(r'$X_1$')
plt.ylabel(r'$X_2$')
plt.legend()
plt.show()
```

## Fazendo testes iniciais com os modelos de redes neurais

```python
xa_ps, x2_ps, w_ps = ModelsRNA.perceptron_simples(data.X, data.Y, .01, 1e3)

xa_ada, x2_ada, w_ada = ModelsRNA.adaptive_linear_element(
    data.X, data.Y, .001, 1e8, 1e-6)

# xa_pm, x2_pm, w_pm = ModelsRNA.perceptron_multipo(
#     data.W, data.Y, .01, 1e3, 2, 4, 5, .001)

print("XA Percp Simpl")
display(xa_ps)

print("X2 Percp Simpl")
display(x2_ps)

print("W Percp Simpl")
display(w_ps)

print("XA ADAline")
display(xa_ada)

print("X2 ADAline")
display(x2_ada)

print("W ADAline")
display(w_ada)

# print("XA Percp Multpl")
# display(xa_pm)

# print("X2 Percp Multpl")
# display(x2_pm)

# print("W Percp Multpl")
# display(w_pm)
```
